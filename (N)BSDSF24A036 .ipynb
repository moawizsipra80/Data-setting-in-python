```python
traffic_logs = [
    "id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
    "id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
    "id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
    "id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
    "id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth"
]

converted_speeds = []
for log in traffic_logs:
    parts = log.split(",")
    for p in parts:
        key, value = p.split(":", 1)
        if key.strip() == "speed":

            conv_speed = int(value.strip()) * 60 / 3600
            converted_speeds.append(conv_speed)

print("Converted speeds:", converted_speeds)
def maximum_speed(speed_list):
    speeds_over_80 = [s for s in speed_list if s > 80]
    if speeds_over_80:
        return max(speeds_over_80)
    else:
        return None

print("The maximum converted speed (above 80) is:", maximum_speed(converted_speeds))
```

```python
traffic_logs = [
    "id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
    "id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
    "id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
    "id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
    "id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth"
]
parsed_logs = []
for log in traffic_logs:
    parts = log.split(",")
    rec = {}
    for p in parts:
        key, value = p.split(":", 1)
        key = key.strip()
        value = value.strip()
        if key == "speed":
            rec[key] = int(value)
        elif key == "violations":
            inner = value[1:-1]    
            if inner and  inner !="None":
             rec[key] = []
            else:
                rec[key] = [inner]
        else:
            rec[key] = value
    parsed_logs.append(rec)

vehicle_summary = {}
for rec in parsed_logs:
    vtype = rec["vehicle"]
    if vtype not in vehicle_summary:
        vehicle_summary[vtype] = {
            "count": 0,
            "speed_sum": 0,
            "violations_count": 0
        }
    vehicle_summary[vtype]["count"] += 1
    vehicle_summary[vtype]["speed_sum"] += rec["speed"]
    vehicle_summary[vtype]["violations_count"] += len(rec["violations"])

for vtype, info in vehicle_summary.items():
    info["avg_speed"] = info["speed_sum"] / info["count"]
    del info["speed_sum"]

print("Vehicle category summary:")
for vtype, info in vehicle_summary.items():
    print(vtype, "=", info)
```

```python
traffic_logs = [
    "id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
    "id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
    "id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
    "id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
    "id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth"
]
def time_window(time_str):
    hour=int(time_str.split(":")[0])
    if 5 <= hour <= 11:
        return "Morning"
    elif 12 <= hour <= 16:
        return "Afternoon"
    elif 17 <= hour <= 20:
        return "Evening"
    else:
        return "Night"  
tt=[]
for log in traffic_logs:
    parts = log.split(",")
    rec={}
    for p in parts:
        key,value=p.split(":",1)
        key=key.strip()
        value=value.strip()
        rec[key]=value
    rec["time_window"] =time_window(rec["time"])  
    tt.append(rec)
for r in tt:
    print(r["time"], "=", r["time_window"])
    
```

```python
from collections import Counter   # For finding most common vehicle

traffic_logs = [
    "id:501,zone:A1,vehicle:Car,speed:62,time:08:30,violations:[None],status:Smooth",
    "id:502,zone:A1,vehicle:Bike,speed:85,time:09:10,violations:[Helmet],status:Busy",
    "id:503,zone:B2,vehicle:Bus,speed:45,time:17:25,violations:[None],status:Smooth",
    "id:504,zone:C3,vehicle:Car,speed:110,time:14:15,violations:[Overspeed],status:Congested",
    "id:505,zone:A1,vehicle:Truck,speed:40,time:18:50,violations:[None],status:Smooth"
]

zones = {}

for log in traffic_logs:
    # Saare info extract karo
    parts = log.split(",")
    zone = ''
    speed = 0
    vehicle = ''
    violations = []
    for p in parts:
        key, value = p.split(":", 1)
        key = key.strip()
        value = value.strip()
        if key == "zone":
            zone = value
        elif key == "speed":
            speed = int(value)
        elif key == "vehicle":
            vehicle = value
        elif key == "violations":
            v = value[1:-1]
            violations = [] if v == "None" else [v]
   
    if zone not in zones:
        zones[zone] = {"count": 0, "speed_sum": 0, "violations_total": 0, "vehicles": []}
    zones[zone]["count"] += 1
    zones[zone]["speed_sum"] += speed
    zones[zone]["violations_total"] += len(violations)
    zones[zone]["vehicles"].append(vehicle)
for z, data in zones.items():
    avg_speed = data["speed_sum"] / data["count"]
    most_common_vehicle = Counter(data["vehicles"]).most_common(1)[0][0]
    if data["violations_total"] == 0:
        safety = "Safe"
    elif data["violations_total"] == 1:
        safety = "Moderate"
    else:
        safety = "Risky"
    print(f"Zone: {z}")
    print(f"  Vehicles: {data['count']}")
    print(f"  Avg speed: {avg_speed}")
    print(f"  Total violations: {data['violations_total']}")
    print(f"  Most common vehicle: {most_common_vehicle}")
    print(f"  Safety: {safety}\n")
```
